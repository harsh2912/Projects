```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import cv2
from sklearn.cluster import KMeans
```

```python
im = cv2.imread('./image.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

'''r, c = im.shape[:2]
out_r = 120
im = cv2.resize(im, (int(out_r*float(c)/r), out_r)) #resizing the image to the shape proportional to original
'''
pixels = im.reshape((-1, 3))

print (pixels.shape)
plt.axis('off')
plt.imshow(im)
plt.show()
```

```python
# for top 10 dominant colors
km = KMeans(n_clusters=8)
km.fit(pixels)
```

```python
colors = np.asarray(km.cluster_centers_, dtype='uint8')
```

```python
per = np.asarray(np.unique(km.labels_, return_counts=True)[1], dtype='float32')
per = per / pixels.shape[0]
per
```

```python
plt.figure(0,figsize=(20,10))
for ix in range(colors.shape[0]):
    patch = np.ones((20, 20, 3))
    patch[:, :, :] = colors[ix]
    patch=patch.astype('uint8')
    plt.subplot(1, colors.shape[0], ix+1)
    plt.axis("off")
    plt.imshow(patch)
```

```python
dom = [[per[ix], colors[ix]] for ix in range(km.n_clusters)]
DOM = sorted(dom, key=lambda z: z[0], reverse=True)
```

```python
#plotting based on their percentage of pixels

plt.figure(0,figsize=(20,20))
plt.axis('off')

patch = np.zeros((50, 500, 3))

start = 0
for cx in range(km.n_clusters):
    width = int(DOM[cx][0] * patch.shape[1])
    end = start + width
    patch[:, start:end, :] = DOM[cx][1]
    patch=patch.astype('uint8')
    start = end

plt.imshow(patch)
plt.show()
```

### Regenerating Image using top 8 dominant colors only

```python
dictionary=dict((i,c) for i,c in zip(np.arange(10),km.cluster_centers_.astype('uint8')))
predicted_image=np.zeros((km.labels_.shape[0],3))
for i in range(predicted_image.shape[0]):
    predicted_image[i]=dictionary[km.labels_[i]]

predicted_image=predicted_image.reshape(im.shape).astype('uint8')
plt.figure(0,figsize=(20,20))
plt.subplot(1,2,1)
plt.title('Image with 8 dominant colors')
plt.imshow(predicted_image)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Original Image')
plt.imshow(im)
plt.axis('off')
plt.show()
```
